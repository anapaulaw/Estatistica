```python
import random
import matplotlib.pyplot as plt
from collections import Counter

def r_binom(N, n, p):
    amostras = []
    for _ in range(N):
        sucesso = 0
        for _ in range(n):
            u = random.uniform(0, 1)
            if u < p:
                sucesso += 1
        amostras.append(sucesso)
    return amostras

def main():
    print("Sistema de Geração de Amostras Binomiais")

    try:
        N = int(input("Digite o número de repetições (N): "))
        n = int(input("Digite o número de ensaios por repetição (n): "))
        p = float(input("Digite a probabilidade de sucesso (p, entre 0 e 1): "))

        if not (0 <= p <= 1):
            print("A probabilidade deve estar entre 0 e 1.")
            return

    except ValueError:
        print("Por favor, insira valores válidos.")
        return


    amostras = r_binom(N, n, p)

    print("\nPrimeiros valores da amostra:")
    print(amostras[:10])


    contagem = Counter(amostras)
    total = sum(contagem.values())
    proporcoes = {k: v / total for k, v in sorted(contagem.items())}


    plt.bar(proporcoes.keys(), proporcoes.values(), color='lightgreen', edgecolor='black')
    plt.xlabel('Número de Sucessos')
    plt.ylabel('Proporção')
    plt.title(f'Distribuição Binomial Simulada (n={n}, p={p}, N={N})')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

if __name__ == "__main__":
    main()
```
